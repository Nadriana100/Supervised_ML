```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('2019loans.csv'))
test_df = pd.read_csv(Path('2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
# pd.get_dummies()
train_df["y"]= pd.get_dummies(train_df["loan_status"],drop_first=True ).rename(columns={"low_risk":"y"})
```

```python
train_df.drop(columns=["loan_status", "Unnamed: 0", "index"], inplace=True)
```

```python
# pd.get_dummies()
test_df["y"]= pd.get_dummies(test_df["loan_status"],drop_first=True ).rename(columns={"low_risk":"y"})
```

```python
test_df.drop(columns=["loan_status", "Unnamed: 0", "index"], inplace=True)
```

```python
train_df= pd.concat([train_df.select_dtypes(exclude="object"),pd.get_dummies(train_df.select_dtypes(include="object"))], axis=1)
```

```python
test_df= pd.concat([test_df.select_dtypes(exclude="object"),pd.get_dummies(test_df.select_dtypes(include="object"))], axis=1)
```

```python
for x in train_df.columns: 
  if x not in test_df.columns:
    print(x)
    test_df[x]=0
```

```python
test_df.shape
```

```python
y_train=train_df["y"]
X_train=train_df.drop(columns=["y"])
```

```python
y_test=test_df["y"]
X_test=test_df.drop(columns=["y"])
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=4000).fit(X_train, y_train)
clf.score(X_test, y_test)

print(f"Training Data Score Logistic Regression: {clf.score(X_train, y_train)}")
print(f"Testing Data Score Logistic Regression: {clf.score(X_test, y_test)}")
```

## Logistic Regression model on the unscaled data
I got a better Testing Score using max_inter=4000 (maximum number of iterations taken for the solvers to converge) , having a Testing Data Score: 0.5776265418970651. I got 0.51.... without max_iner 

```python
# Train a Random Forest Classifier model and print the model score

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)

print(f"Training Data Score Random Forest Clf: {clf.score(X_train, y_train)}")
print(f"Testing Data Score Random Forest Clf: {clf.score(X_test, y_test)}")
```

# We got a better result on our TESTING SCORE using the Train Random Forest Classifier, around 10% better result, since fits a number of decision tree classifiers on various sub-samples.

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
# Train the Logistic Regression model on the scaled data and print the model score

clf = LogisticRegression(max_iter=4000).fit(X_train_scaled, y_train)
clf.score(X_test_scaled, y_test)

print(f"Training Data Score Log. Regression SCALE DATA: {clf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score Log. Regression SCALE DATA: {clf.score(X_test_scaled, y_test)}")
```

## Using the SCALED DATA the Logistic Regression have aroun 20% better result than getting the Logostic Regression without scale the data. 

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
clf.fit(X_train_scaled, y_train)
clf.score(X_test_scaled, y_test)

print(f"Training Data Score Random Forest SCALE DATA: {clf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score Random Forest SCALE DATA: {clf.score(X_test_scaled, y_test)}")
```
